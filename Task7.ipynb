```python
#увеличим дефолтный размер графиков
from pylab import rcParams
rcParams['figure.figsize'] = 12,9


# импортируем нужные библиотеки
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
Players = pd.read_csv("Players.csv")
Stats = pd.read_csv("Seasons_Stats.csv")
```

```python
# График количество игроков по годам рождения

filter_data = Players.groupby('born').agg('count')['Player'].reset_index()
plt.bar(filter_data['born'], filter_data['Player'])

# Больше всего игроков 1970 года рождения
```

```python
# Количество очков набранных A.J. Price в каждом году выступо\лений в НБА

filter_data = Stats[Stats['Player'] == 'A.J. Price']
filter_data['FG']
plt.plot(filter_data['Year'],filter_data['PTS'])

# Наибольшее количество очков было набрано в 2013 году, также количество набранных очков падало важдом следующим году исключая 2013.
```

```python
# Гистограмма с количество набранных очков Aaron McKie

filter_data = Stats[Stats['Player'] == 'Aaron McKie']
filter_data.PTS.hist(bins = 7)

# Aaron McKie только в трех сезонах набрад больше 700 очков
```

```python
# Распределение количества очков

filter_data = Stats[Stats['PTS'] >= 0]
sns.distplot(filter_data.PTS)

# Меньше 0.05% набрали больше 1000 очков за сезон
```

```python
# Связь между попытками трех очковых бросков и набранными очками

filter_data = Stats[Stats['3P'] >= 0]

sns.jointplot(x='3PA', y='3P', 
              data=filter_data, kind='scatter')

# Чем больше попыток трех очковых бросков тем больше набранных очков
```
